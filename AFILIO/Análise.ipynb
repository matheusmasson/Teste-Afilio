```python
rawds['dia'] = rawds["Data da receita"].str.split("/").apply(lambda x: x[0])
rawds["mes"] = rawds["Data da receita"].str.split("/").apply(lambda x: x[1])
rawds["ano"] = rawds["Data da receita"].str.split("/").apply(lambda x: x[2][:4])
```

```python
rawds["Valor receita"] = rawds["Valor receita"].str.replace(',', '.').apply(float)
```

```python
# Cria um banco de dados sqlite para facilitar as consultas já com os dados tratados

conn = sqlite3.connect("/home/matheus/Documents/AFILIO/dbAfilio.db")
cur = conn.cursor()

rawds.to_sql('dataset', conn, if_exists='replace', index=False)
```
